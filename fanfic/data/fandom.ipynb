```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("fandomRanking.csv") 
df
```

```python
# Group by Media and sum New Works
media_totals = df.groupby("Media")["New Works"].sum().reset_index()
```

```python
# Sort from highest to lowest
media_totals = media_totals.sort_values(by="New Works", ascending=False)
```
